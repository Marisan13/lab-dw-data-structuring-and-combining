In this challenge, we will be working with the customer data from an insurance company, as we did in the two previous labs. The data can be found here:

But this time, we got new data, which can be found in the following 2 CSV files located at the following links:
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2_clean.csv
- https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3_clean.csv

Combine the data from the three dataframes into a single dataframe, named "customer_data", using appropriate merging, concatenating, and joining techniques.

Verify that the customer_data dataframe contains all the rows and columns from the three original dataframes.

```python
import pandas as pd

#pd.options.display.max_columns = None
#pd.options.display.max_rows = None

url_1 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv'
url_2 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2_clean.csv'
url_3 = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3_clean.csv'

df_1= pd.read_csv(url_1)
df_2= pd.read_csv(url_2)
df_3= pd.read_csv(url_3)

display(df_1.head(15))
display(df_2.head(15))
display(df_3.head(15))
```

```python
print(df_1.shape)
print(df_2.shape)
print(df_3.shape)
```

First confirmed the difference between the three tables
<br>
- df_2 and df_3 are totally the same.
- df_1 is covered in df_2 (as well as df_3).  >>> what is the point to combine???

```python
# Confirming the columns of the three tables - all equal

df_1.columns.equals(df_2.columns) and df_2.columns.equals(df_3.columns)
```

```python
# Confirming whether df_1 is covered in df_2 or not - Yes

# Filling null cellls with 0, as "in" does not work with null cells when comparing the two tables.
df_1_wo_null = df_1.fillna(0)
df_2_wo_null = df_2.fillna(0)

# Checking whether df_1 data is found in df_2 or not row by row
all(df_1_wo_null.iloc[i].to_list() in df_2_wo_null.to_numpy().tolist() for i in range(len(df_1_wo_null)))
```

```python
# Confirming whether df_2 and df_3 are same or not -  Yes. 

df_2.equals(df_3)
```

Confirming the shape of df_2 or df_3 but without duplicates

```python
df_2_no_duplicates = df_2.drop_duplicates(keep="first")
df_2_no_duplicates.shape
```

Combining the tables

- Concatenating the three tables and dropping the duplicates

```python
# Concatenating the three tables
customer_data = pd.concat([df_1, df_2, df_3])

# Dropping the duplicates keeping the first value
customer_data = customer_data.drop_duplicates(keep="first")
```

```python
# Confirming that the number of rows and columns equals to that of df_2 (and df_3) without duplicates.

customer_data.shape
```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv'

df= pd.read_csv(url)

print(df.shape)
display(df.head())
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
# Using pivot

df.pivot_table(index="sales_channel", values="monthly_premium_auto", aggfunc="sum")
```

```python
# Using groupby and agg

df.groupby(["sales_channel"]).agg({"monthly_premium_auto": ["sum"]})
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
# Creating pibot table

average_clv_gender = df.groupby(["gender", "education"]).agg({"customer_lifetime_value": ["mean"]})
average_clv_gender
```

```python
# Creating bar chart for analysis

import matplotlib.pyplot as plt

average_clv_gender = df.groupby(["gender", "education"]).agg({"customer_lifetime_value": ["mean"]})

# Reset column names for clarity
average_clv_gender.columns = ["Average_CLV"]

# Plot the bar chart with different colors and data labels
ax = average_clv_gender.plot(kind='bar', figsize=(10, 6))
ax.set_xlabel('GenderEducation')
ax.set_ylabel('Average Customer Lifetime Value')
ax.set_title('CLV by Gender/Education')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Set the y-label above 6000
ax.set_ylim(6000, 9000)

plt.show()
```

**Insights**
<br>
- Females of "High School or Below" group are found with the highest CLV. Reasons may include the possibility of their relatively early timing of marriage and starting a family, the group's behavioral tendencies (e.g., preference for stability), etc.
- Across both genders, the group with Doctor degree has the lowest CLV, and one possible reason may be their relatively late timing of insurance enrollment.
